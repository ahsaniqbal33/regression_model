### Regression

```python
import statsmodels.api as sm
```

```python
import pandas as pd
sales_table = pd.read_excel(r'E:\Python Bootcamp\Adventure Works 2020.xlsx', sheet_name = 'Sales')
```

```python
sales_table
```

```python
 sales_table.columns
```

```python
# Dependent variable ----> Sales Amount
# Independent Variable ----> Total Product Cost, order quantity, tax amount, Unit price, Sales Territory key
```

```python
dependent_variable = sales_table['SalesAmount']
independent_variables = sales_table[['TotalProductCost','OrderQuantity', 'UnitPrice', 'SalesTerritoryKey','TaxAmt' ]]
```

```python
dependent_variable
```

```python
independent_variables
```

```python
regression_model = sm.OLS(dependent_variable, independent_variables)
results = regression_model.fit()
results.summary()                       # OLS means ordinary least square aka linear regression
```
